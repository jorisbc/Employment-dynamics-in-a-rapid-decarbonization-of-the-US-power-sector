```python
# Make a dictionary with all attributes
node_dicts = df_nodes.to_dict('records')

# attributes to be considered
attribute_columns = ['TOT_EMP', 'A_MEAN', 'TOT_EMP_log', 'shock_before_2034', \
    'shock_after_2034', 'shock_before_2034_norm', 'shock_after_2034_norm', \
    'pool_shock_inbefore_norm', 'pool_shock_inafter_norm',\
    'pool_shock_outbefore_norm','pool_shock_outafter_norm','archetype', \
        'classification', 'Polygon', 'polygon_archetype', 'Macro_class']

# create a dictionary mapping OCC_source values to node attributes
node_attributes = {}
for node_dict in node_dicts:
    node_name = node_dict['O*NET-SOC Code']
    node_attributes[node_name] = {
        attribute_name: node_dict[attribute_name] for attribute_name \
            in attribute_columns
    }
    
########
# Edgelist
########
occs_edgelist = set(df_edgelist['OCC_source'])
occs_arch = set(df_nodes['O*NET-SOC Code'])

print('in edgelist not in arch ', occs_edgelist.difference(occs_arch))
print('in arch not in edglist ', occs_arch.difference(occs_edgelist))
```

```python


df_edgelist = df_edgelist[['OCC_source', 'OCC_target', transition_type]]
df_edgelist = df_edgelist.rename({transition_type: 'weight'}, \
    axis='columns')

df_edgelist.head()

G = nx.from_pandas_edgelist(df_edgelist, source='OCC_source', \
    target='OCC_target', edge_attr='weight', create_using=nx.DiGraph())

# set node attributes
nx.set_node_attributes(G, node_attributes)

# export graph 
#nx.write_gexf(G, path_gephi + 'occmob_cc_mergeweight_attributes.gexf')
nx.write_gexf(G, path_gephi + 'cc_edgelist.gexf'.format(n_top))


# remove self loops
G.remove_edges_from(nx.selfloop_edges(G))

zero_edges = [(a,b) for a,b,c in G.edges(data=True) if c['weight'] == 0]
print('number of zero edges ', len(zero_edges))

# remove zero edges
for edge in zero_edges:
    G.remove_edge(edge[0], edge[1])

nx.write_gexf(G, path_gephi + 'cc_edgelist_2.gexf'.format(n_top))

G_neg = G.copy()

for edge in G_neg.edges():
    G_neg.edges[edge]['weight'] = -G_neg.edges[edge]['weight']
```

```python
UG = G_neg.to_undirected()
for node in G_neg:
    for ngbr in nx.neighbors(G_neg, node):
        if node in nx.neighbors(G_neg, ngbr):
            UG.edges[node, ngbr]['weight'] = (
                G_neg.edges[node, ngbr]['weight'] + G_neg.edges[ngbr, node]['weight']
            ) / 2
```

```python
# minimum spanning tree
T = nx.minimum_spanning_tree(UG)

T = T.to_directed()

# reset weights to directed graph and positive
for edge in T.edges():
    try:
        T.edges[edge]['weight'] = G.edges[edge]['weight']
    except KeyError:
        T.edges[edge]['weight'] = 0


nx.set_node_attributes(T, node_attributes)

mean_edge_weight = np.mean([G.edges[edge]['weight'] for edge in G.edges()])
median_edge_weight = np.median([G.edges[edge]['weight'] for edge in G.edges()])
std_edge_weight = np.std([G.edges[edge]['weight'] for edge in G.edges()])

# extra edges = edges with weight > mean + 1 std
extra_edges = [edge for edge in G.edges() if G.edges[edge]['weight'] > \
    median_edge_weight]# + 0.5*std_edge_weight]

# len(extra_edges) # 0
print('number of extra edges ', len(extra_edges))
```

```python
# total number of edges in G
print('number of edges in G ', len(G.edges()))
```

```python
# add extra edges to T
for edge in extra_edges:
    T.add_edge(edge[0], edge[1])
    T.edges[edge]['weight'] = G.edges[edge]['weight']
```

```python
zero_edges = [(a,b) for a,b,c in T.edges(data=True) if c['weight'] == 0]
print('number of zero edges ', len(zero_edges))

# remove zero edges
for edge in zero_edges:
    T.remove_edge(edge[0], edge[1])
```
